```python
pop_data = []
for pop in range (1,10000):
    pop_data.append(np.random.randint(1, 101))
pop_table = pd.DataFrame({"data": pop_data})
plt.figure(figsize=[3,3])
sns.kdeplot(x="data", data=pop_table)
plt.show()
```

```python
sampled_mean = []
for i in range (1,20):
    sampled_data = []
    for j in range (1,500):
        sampled_data.append(np.random.choice(pop_data))
    sampled_mean.append(np.mean(sampled_data))

sampled_mean_table = pd.DataFrame({"data_mean": sampled_mean})
plt.figure(figsize=[3,3])
sns.kdeplot(x="data_mean", data=sampled_mean_table)
plt.show()
```

```python
np.mean(sampled_mean), np.mean(pop_data) # notice that sampled mean and the mean of original data are almost same, hence CLT is verified
```
